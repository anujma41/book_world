```python
# Try skipping bad lines with on_bad_lines='skip'
rating = pd.read_csv("C:/Users/anujm/Downloads/BOOK RECOMMDED SYSTEM/BX-Book-Ratings.csv/BX-Book-Ratings.csv", sep=";", encoding='latin1', on_bad_lines='skip')

```

```python
# Assuming the actual ISBN column in the 'books' DataFrame is named 'book_id'
rating_with_books = rating.merge(books, on="isbn")
```

```python
num_rating = rating_with_books.groupby('Title')['Book-Rating'].count().reset_index()
print(num_rating)
num_rating.rename(columns={"Book-Rating":"num_of_rating"},inplace=True)
```

```python
final_rating = rating_with_books.merge(num_rating, on="Title")
final_rating.head(5)
```

```python
final_rating=final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.drop_duplicates(['User-ID','Title'],inplace=True)
final_rating.shape
```

```python
final_rating.head()
```

```python

book_pivot = final_rating.pivot_table(columns='User-ID', index='Title', values='Book-Rating')
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
pip install scipy
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
book_sparse
```

```python
pip install scikit-learn
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
# Define the number of neighbors separately
n_neighbors = 6

# Calculate the distance and suggestions
distance, suggestion = model.kneighbors(book_pivot.iloc[270, :].values.reshape(1, -1), n_neighbors=n_neighbors)
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
books_name = book_pivot.index.tolist()
books_name
```

```python
import pickle

# Save the model and data objects to the 'artifacts' directory
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl', 'wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl', 'wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot.pkl', 'wb'))
```

```python
def recommend_book(books_name):
    # Print the input book name for debugging
    print(f"Input book name: '{books_name}'")

    # Check if books_name exists in book_pivot.index
    if books_name not in book_pivot.index:
        print(f"Book '{books_name}' not found in the dataset.")
        return         
    
    # Get the book ID
    book_id = np.where(book_pivot.index == books_name)[0][0]
    print(f"Book ID for '{books_name}': {book_id}")

    # Get distance and suggestions from the model
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    
    # Print distance and suggestion for debugging
    print(f"Distance: {distance}")
    print(f"Suggestion indices: {suggestion}")

    # Loop through the suggestions and print book names
    for i in range(1, len(suggestion[0])):  # Skip the first one as it's the same book
        suggested_book = book_pivot.index[suggestion[0][i]]
        print(suggested_book)

# Assuming 'books_name' refers to a single book title (e.g., from a selectbox or input)
selected_books = "The Notebook"  # Example selected book
recommend_book(selected_books)
```
